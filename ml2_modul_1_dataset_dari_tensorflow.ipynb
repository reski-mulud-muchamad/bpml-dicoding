```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(gambar_latih, label_latih), (gambar_testing, label_testing) = mnist.load_data()
```

```python
np.set_printoptions(linewidth=200)
plt.imshow(gambar_latih[0])
print(label_latih[0])
```

```python
# normalisasi
gambar_latih = gambar_latih / 255.0
gambar_testing = gambar_testing / 255.0
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28, 28)),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(gambar_latih, label_latih, batch_size=128, epochs=5)
```
