```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
tokenizer = Tokenizer(num_words=15, oov_token='-')
text = ['Saya suka programming!',
        'Programming sangat menyenangkan!',
        'Machine Learning berbeda dengan pemrograman konvensional']
```

```python
tokenizer.fit_on_texts(text)
```

```python
sequence = tokenizer.texts_to_sequences(text)
```

```python
print(tokenizer.word_index)
```

```python
print(tokenizer.texts_to_sequences([text[0]]))
print(tokenizer.texts_to_sequences([text[1]]))
print(tokenizer.texts_to_sequences([text[2]]))
```

```python
sequence_samapanjang = pad_sequences(sequence)
print(sequence_samapanjang)
```

```python
sequence_samapanjang = pad_sequences(sequences=sequence,
                                     padding='post')
print(sequence_samapanjang)
```
